```python
import mne
import os.path as op
%matplotlib inline

subject = 'A0148'
path = '/Users/teon/Google Drive/E-MEG/data/%s/mne/' % subject
plot_path = '/Applications/packages/E-MEG/output/results/%s/meg' % subject
# hp: 0.5 Hz, lp: 40 Hz
raw_file = op.join(path, '%s_OLDT_calm_filtered-raw.fif' % subject)
ica_file = op.join(path, raw_file[:-7] + 'ica.fif')
event_file = op.join(path, '%s_OLDT-eve.txt' % subject)
```

```python
evts = mne.read_events(event_file)
raw = mne.io.Raw(raw_file, preload=True)
```

```python
epochs = mne.Epochs(raw, evts, None, tmin=-.1, tmax=.5, decim=4, baseline=None, verbose=False)
evoked = epochs.average()
```

```python
evoked.plot();
```

```python
cov = mne.compute_covariance(epochs[::10], tmin=-.1, tmax=0, method='auto', verbose=False)
```

```python
evoked.plot_white(cov);
```

```python
if op.exists(ica_file):
    ica = mne.preprocessing.read_ica(ica_file)
    ica.fit(raw, decim=10)
else:
    ica = mne.preprocessing.ICA(.9, random_state=42, method='infomax')
    ica.fit(raw, decim=10)
```

```python
ica.plot_components();
```

```python
ics = ica.get_sources(raw)
```

```python
epochs = mne.Epochs(ics, evts, None, tmin=-.1, tmax=.5, decim=4, baseline=None, verbose=False)
picks = range(len(epochs.info['ch_names']))
evoked = epochs.average(picks)
```

```python
evoked.plot();
```

```python
for ic in picks:
    evoked.plot([ic], titles={'misc': 'IC %d' % ic}, ylim = dict(misc=[-1, 1]));
```
